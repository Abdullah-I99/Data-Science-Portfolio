```python
#import necessary libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load cleaned data for EDA
eda_path = "DataCoSupplyChainDataset_cleaned.csv"
df_eda = pd.read_csv(eda_path)

print(df_eda.shape)
df_eda.head()
```

```python
# Get basic info and statistics
df_eda.info()
df_eda.describe()
```

```python
# Visualize the distribution of delivery risk (lat vs on time)
plt.figure(figsize=(5,4))
sns.countplot(x="Late_delivery_risk", data=df_eda)
plt.title("Late Delivery Risk Distribution (0 = On Time, 1 = Late)")
plt.xlabel("Late_delivery_risk")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Visualize shipping delay days vs late delivery risk
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Late_delivery_risk",
    y="shipping_delay_days",
    data=df_eda
)
plt.title("Shipping Delay vs Late Delivery Risk")
plt.xlabel("Late_delivery_risk (0 = On Time, 1 = Late)")
plt.ylabel("shipping_delay_days (real - scheduled)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="Days for shipment (scheduled)",
    y="Days for shipping (real)",
    data=df_eda,
    alpha=0.3
)
plt.title("Real vs Scheduled Shipping Days")
plt.xlabel("Days for shipment (scheduled)")
plt.ylabel("Days for shipping (real)")
plt.tight_layout()
plt.show()
```

```python
# Visualize average shipping delay by shipping mode
late_by_mode = df_eda.groupby("Shipping Mode")["shipping_delay_days"].mean()

plt.figure(figsize=(8,4))
late_by_mode.sort_values().plot(kind="bar")
plt.title("Average Shipping Delay by Shipping Mode")
plt.ylabel("Average shipping_delay_days")
plt.xlabel("Shipping Mode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
