```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding="latin1")

# quick look at the data
print(df.shape)
df.head()

```

```python
# quick look basic info
df.info()
df.describe(include='all')
```

```python
# Select relevant columns
cols = [
    "Late_delivery_risk",
    "Delivery Status",
    "Days for shipping (real)",
    "Days for shipment (scheduled)",
    "order date (DateOrders)",
    "shipping date (DateOrders)",
    "Shipping Mode",
    "Order Region",
    "Order Country",
    "Order Item Quantity",
    "Sales",
    "Order Item Total"
]

df = df[cols].copy()
df.head()
```

```python
# Convert date columns to datetime format
df["order date (DateOrders)"] = pd.to_datetime(
    df["order date (DateOrders)"], errors="coerce"
)
df["shipping date (DateOrders)"] = pd.to_datetime(
    df["shipping date (DateOrders)"], errors="coerce"
)

df[["order date (DateOrders)", "shipping date (DateOrders)"]].head()
```

```python
# Calculate the difference in days between order date and shipping date
df["order_to_ship_days"] = (
    df["shipping date (DateOrders)"] - df["order date (DateOrders)"]
).dt.days
```

```python
# Calculate the number of days delayed in shipping
df["shipping_delay_days"] = (
    df["Days for shipping (real)"] - df["Days for shipment (scheduled)"]
)
```

```python
# Drop original date columns as irrelevant for any analaysis now
df = df.drop(columns=[
    "order date (DateOrders)",
    "shipping date (DateOrders)"
])
```

```python
#check for missing values
df.isnull().sum()
```

```python
# save cleaned data

clean_path = "DataCoSupplyChainDataset_cleaned.csv"
df.to_csv(clean_path, index=False)
clean_path
```
